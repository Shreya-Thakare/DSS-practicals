```python
#importing the basic library
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=x*2
```

```python
y
```

Line Chart

```python
plt.plot(x,y)
plt.title("Line chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Bar Chart

```python
plt.bar(x,y)
plt.title("Bar chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.bar(x,y,color="purple")
plt.title("Bar chart")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Scatter Chart

```python
a=(1,5,8,6,3,7,9,4)
b=(10,55,21,64,85,33,44,28)
plt.scatter(a,b)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Histogram

```python
H=[1,1,1,5,5,8,6,6,7,7,9,9,2,2,4,4,4,2,2,2]
```
